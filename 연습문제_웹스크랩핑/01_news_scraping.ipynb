```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/economy'

response = requests.get(url)
response.encoding = 'utf-8'

print(url)
print(type(response))
print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')

news_list = soup.select('ul.list_newsheadline2 li')
print(type(news_list), len(news_list))

for li_tag in news_list:
    a_tag = li_tag.find('a')
    if a_tag:
        link = a_tag['href']
        strong_tag = li_tag.select_one('strong.tit_txt')
        if strong_tag:
            title = strong_tag.text.strip()
            print(link)
            print(title)
```

```python
import requests
from bs4 import BeautifulSoup

def print_news(section_name):
    section_dict = {
        '기후/환경': 'climate', '사회': 'society', '경제': 'economy', '정치': 'politics',
        '국제': 'world', '문화': 'culture', '생활': 'life', 'IT/과학': 'tech', '인물': 'people'
    }
    
    section_path = section_dict.get(section_name)
    if not section_path:
        return
        
    url = f'https://news.daum.net/{section_path}'
    
    response = requests.get(url)
    response.encoding = 'utf-8'
    
    print(f"======> {url} {section_name} 뉴스 <======")
    
    soup = BeautifulSoup(response.text, 'html.parser')
    news_list = soup.select('ul.list_newsheadline2 li')
    
    for li_tag in news_list:
        a_tag = li_tag.find('a')
        if a_tag:
            link = a_tag['href']
            strong_tag = li_tag.select_one('strong.tit_txt')
            if strong_tag:
                title = strong_tag.text.strip()
                print(link)
                print(title)

print_news('경제')
print_news('사회')
```
