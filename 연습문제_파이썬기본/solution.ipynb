```python
def fact_rec(n): return 1 if n <= 1 else n * fact_rec(n-1)
def fact_iter(n):
    res = 1
    for i in range(1, n+1): res *= i
    return res
print(f"5! (재귀) = {fact_rec(5)}")
print(f"5! (반복) = {fact_iter(5)}")
```

```python
import statistics
def get_stats(nums):
    return sum(nums)/len(nums), max(nums), min(nums), statistics.stdev(nums)

nums = [10, 20, 30, 40, 50]
avg, mx, mn, sd = get_stats(nums)
print(f"평균: {avg}, 최댓값: {mx}, 최솟값: {mn}, 표준편차: {sd:.2f}")
```
